```python
import requests
from config import key
from pprint import pprint
import time
import pandas as pd
key
```

```python
tier = ["GRANDMASTER", "CHALLENGER"]
division = ["I", "II", "III", "IV"]
page = 1
combined_data = pd.DataFrame()
emptypage = False

while emptypage == False:
    url = f"https://na1.api.riotgames.com/lol/league-exp/v4/entries/RANKED_SOLO_5x5/"\
          f"PLATINUM/I?page={page}&api_key={key}"
    data = requests.get(url).json()
    if data == []:
        print(f"You've reached empty page, the last page is {page -1}")
        emptypage = True
    if type(data) is dict:
        print(data["status"]["message"])
        break
    data_df = pd.DataFrame(data)
    combined_data = combined_data.append(data_df, ignore_index=True)
    page += 1
    if page % 10 == 0:
        print ("10 pages done, now sleeping for 15 seconds to aovid reaching limit")
        time.sleep(15)

combined_data
```
