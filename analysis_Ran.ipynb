```python
import requests
from config import key
from pprint import pprint
import time
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# To get all player information from League-Exp-V4 api.
# And save each rank to csv.

tiers = ["IRON", "BRONZE", "SILVER", "GOLD", "PLATINUM", "DIAMOND","MASTER","GRANDMASTER","CHALLENGER"]
divisions = ["I", "II", "III", "IV"]

for tier in tiers:
    for division in divisions:
        combined_data = pd.DataFrame()
        page = 1
        emptypage = False
        while emptypage == False:
            url = f"https://na1.api.riotgames.com/lol/league-exp/v4/entries/RANKED_SOLO_5x5/"\
                  f"{tier}/{division}?page={page}&api_key={key}"
            data = requests.get(url).json()
            if data == []:
                print(f"You've reached empty page for {tier} {division}, the last page is {page -1}")
                emptypage = True
            if type(data) is dict:
                print(data["status"]["message"])
                break
            data_df = pd.DataFrame(data)
            combined_data = combined_data.append(data_df, ignore_index=True)
            page += 1
            if page % 10 == 0:
                print (f"{tier} {division}: 10 pages done, sleep for 10 seconds to aovid reaching limit")
                time.sleep(10)
        if combined_data.empty == False:
            combined_data.to_csv(f"population_data/{tier}_{division}_population.csv")

```

```python
# Trimming the csv file based on unique summonerID and count the number from each csv
# As well as delete the extra column
# save to that csv file
population_path = "population_data/"
total_count = 0
for file in os.listdir(population_path):
    current_df = pd.read_csv(os.path.join(population_path,file))
    tier_rank = file.split("_")
    del current_df["Unnamed: 0"]
    current_df = current_df[(current_df["tier"] == tier_rank[0]) | (current_df["rank"] == tier_rank[1])]
    current_df = current_df.drop_duplicates(subset=["summonerId"])
    current_df = current_df.reset_index(drop = True)
    count_by_ID = current_df["summonerId"].nunique()
    count_by_row = len(current_df.index)
    if count_by_ID == count_by_row:
        print(f"{tier_rank[0]} {tier_rank[1]} has a total player of {count_by_ID}")
        total_count += count_by_ID
        current_df.to_csv(f"population_data/{tier_rank[0]}_{tier_rank[1]}_population.csv", index=False)
    else:
        print(f"Oops there is mismatch in {tier_rank[0]} {tier_rank[1]}")
print(total_count)
```

```python
# So we have 2144187 total players from all the ranks
# Now we need 30k of total subset population, which is 1.4%
# So I will take 1.4% of player from each rank by pd sample function
total_subset_df = pd.DataFrame(columns = ["leagueId","queueType","tier","rank","summonerId","summonerName","leaguePoints","wins","losses","veteran","inactive","freshBlood","hotStreak"])
population_path = "population_data/"

for file in os.listdir(population_path):
    current_df = pd.read_csv(os.path.join(population_path,file))
    tier_rank = file.split("_")
    count_by_ID = current_df["summonerId"].nunique()
    rank_subset_count = np.ceil(count_by_ID*0.014).astype(int)
    print(f"{tier_rank[0]} {tier_rank[1]} has a subset of {rank_subset_count} players")
    subset_df = current_df.sample(n=rank_subset_count)
    
    # I check if there is summoner ID already sampled in the total subset, becuase
    # looping through all ranks took me 6 hours, some players might moved to other ranks during
    # the time, causing duplicate ID pulled in more than one rank csv.
    
    duplicate_ID = list(set(subset_df["summonerId"]).intersection(set(total_subset_df["summonerId"])))
    # If there is duplicate, append non-duplicate ID
    subset_df = subset_df[~subset_df['summonerId'].isin(duplicate_ID)]
    total_subset_df = total_subset_df.append(subset_df, ignore_index=True)
    
    # If there is duplicate, we are not taking them into total subset, so we might miss a few 
    # input to reach the desired sample size. Creating while loop here to resample whatever 
    # many gap we are missing because of duplicate.
    while len(duplicate_ID) > 0:
        print (f"Ooops there are {len(duplicate_ID)} summoner IDs that are already sampled when"\
               f"picking random subset from {tier_rank[0]} {tier_rank[1]}, try resample "\
               f"{len(duplicate_ID)} summoner IDs")
        resample_df = current_df.sample(n=len(duplicate_ID))
        duplicate_ID = list(set(resample_df['summonerId']).intersection(set(total_subset_df["summonerId"])))
        subset_df = resample_df[~resample_df["summonerID"].isin(duplicate_ID)]
        total_subset_df = total_subset_df.append(subset_df, ignore_index=True)
```

```python
# Demographic distribution among tiers.

tier_demographic = summoner_champion_df.groupby(["tier", "rank"])["leagueId"].count()
tier = ["IRON", "BRONZE", "SILVER", "GOLD", "PLATINUM", "DIAMOND","MASTER","GRANDMASTER","CHALLENGER"]
tier_demographic = tier_demographic.reindex(tier, level=0)
tier_demographic = tier_demographic.sort_index(level=[0,1], ascending=[1,0])
tier_demographic = tier_demographic.reset_index()
tier_demographic.columns = ["Tier", "Rank", "Count"]
sns.barplot(x="Tier", y="Count", hue="Rank", data=tier_demographic)
plt.xticks(rotation=50, ha='right')
plt.title("Demographic Distribtion across all Tiers")
plt.grid()
```

```python
rank_win = summoner_champion_df.groupby('tier')["wins"].mean().astype(int).reset_index()
rank_loss = summoner_champion_df.groupby('tier')["losses"].mean().astype(int).reset_index()
win_loss = rank_win.merge(rank_loss, on="tier", how="left")
win_loss = win_loss.set_index('tier')
tier = ["IRON", "BRONZE", "SILVER", "GOLD", "PLATINUM", "DIAMOND","MASTER","GRANDMASTER","CHALLENGER"]
win_loss = win_loss.reindex(tier)
win_loss = win_loss.reset_index()
win_loss.columns = ["Tier", "Wins", "Losses"]
win_loss["Losses"] = [-x for x in win_loss["Losses"]]

win_std = summoner_champion_df.groupby('tier')["wins"].std()
win_std = win_std.reindex(tier).tolist()
loss_std = summoner_champion_df.groupby('tier')["losses"].std()
loss_std = loss_std.reindex(tier).tolist()

labels = win_loss["Tier"]
wins = win_loss["Wins"]
losses = win_loss["Losses"]

fig, ax = plt.subplots()

ax.bar(labels, wins, width=0.8, yerr = win_std, label='Wins', color='tomato')
ax.bar(labels, losses, width=0.8, yerr = loss_std, bottom=0,label='Losses', color='dodgerblue')

ax.set_ylabel('Average Number')
ax.set_title('Wins vs Losses Number and Ratio Across Tiers')
ax.legend(bbox_to_anchor=(1.24,1))
plt.xticks(rotation=50, ha='right')

rects=ax.patches
labels = (wins/abs(losses)).round(3).tolist()

for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text((rect.get_x() + rect.get_width()/2), 0, label, ha='center', va='center',fontweight='semibold')

plt.grid()
plt.show()
```
